```python
data = {'book' : [],
        'price' : [],
        'UPC' : [],
        'Product Type' : [],
        'Price (excl. tax)' : [],
        'Price (incl. tax)' : [],
        'Tax' : [],
        'Availability' : [],
        'Number of reviews' : [],
        'Description' : [],
        'Stars' : []
        }
for books in bookshelf:
  # collect title of all books
  book_title = books.h3.a["title"]

  # collect book price of all books
  book_price = books.findAll("p", {"class": "price_color"})
  price = book_price[0].text.strip()

  data['book'].append(book_title)
  data['price'].append(price)

  url_book = "http://books.toscrape.com/" + books.h3.a['href']

  result = requests.get(url_book)
  result.encoding = 'UTF-8'
  soup_book = BeautifulSoup(result.text, 'html.parser')

  for tr in soup_book.findAll('table')[0].findAll('tr'):
    data[tr.th.text].append(tr.td.text)

  data['Description'].append(soup_book.findAll('div', {'id' : 'product_description'})[0].nextSibling.nextSibling.text)
  data['Stars'].append(soup_book.find("p", class_ = "star-rating").get("class")[-1])
```

### IMDB

```python
url = "https://www.imdb.com/search/title?count=100&title_type=feature,tv_series&ref_=nv_wl_img_2"
```

```python
page = get(url)
soup = BeautifulSoup(page.content, 'lxml')
```

```python
soup.find("h1", class_="header").text.replace("\n","")
```

```python
def bodyContent(soup):
  content = soup.find(id="main")
  return content.find_all("div", class_="lister-item mode-advanced")
```

```python
movieFrame = bodyContent(soup)
data = {'movieTitle' : [],
    'movieDate' : [],
    'movieRunTime' : [],
    'movieGenre' : [],
    'movieRating' : [],
    'movieScore' : [],
    'movieDescription' : [],
    'movieDirector' : [],
    'movieStars' : [],
    'movieVotes' : [],
    'movieGross' : [],
}


for movie in movieFrame:
  movieFirstLine = movie.find("h3", class_="lister-item-header")
  data['movieTitle'].append(movieFirstLine.find("a").text)
  data['movieDate'].append(re.sub(r"[()]","", movieFirstLine.find_all("span")[-1].text))
  try:
    data['movieRunTime'].append(movie.find("span", class_="runtime").text[:-4])
  except:
    data['movieRunTime'].append(np.nan)
  data['movieGenre'].append(movie.find("span", class_="genre").text.rstrip().replace("\n","").split(","))
  try:
    data['movieRating'].append(movie.find("strong").text)
  except:
    data['movieRating'].append(np.nan)
  try:
    data['movieScore'].append(movie.find("span", class_="metascore unfavorable").text.rstrip())
  except:
    data['movieScore'].append(np.nan)
    
  data['movieDescription'].append(movie.find_all("p", class_="text-muted")[-1].text.lstrip())
  movieCast = movie.find("p", class_="")
  try:
    casts = movieCast.text.replace("\n","").split('|')
    casts = [x.strip() for x in casts]
    casts = [casts[i].replace(j, "") for i,j in enumerate(["Director:", "Stars:"])]
    data['movieDirector'].append(casts[0])
    data['movieStars'].append([x.strip() for x in casts[1].split(",")])
  except:
    casts = movieCast.text.replace("\n","").strip()
    data['movieDirector'].append(np.nan)
    data['movieStars'].append([x.strip() for x in casts.split(",")])

  movieNumbers = movie.find_all("span", attrs={"name": "nv"})

  if len(movieNumbers) == 2:
    data['movieVotes'].append(movieNumbers[0].text)
    data['movieGross'].append(movieNumbers[1].text)
  elif len(movieNumbers) == 1:
    data['movieVotes'].append(movieNumbers[0].text)
    data['movieGross'].append(np.nan)
  else:
    data['movieVotes'].append(np.nan)
    data['movieGross'].append(np.nan)

movieData = pd.DataFrame(data)
```
